```python
"""
City and State are object type, which as we have seen in class, represents strings in Panda.
It makes sense in this case, as these are categorical values in text format.
Temperature is float64 type which indicate a floating point number in Pandas.  
This also makes sense as it allows a precise representation of temperature values, 
which can include decimals, making it suitable for calculations or when exactitude is needed.
"""
```

#### Select the rows where state is New York.

```python
# your code here
temp[temp['State'] == 'New York']
```

#### What is the average temperature of cities in New York?

```python
# your code here

mean_temp_ny = temp[temp['State'] == 'New York']['Temperature'].mean()

print(f"The mean temperature of cities in New York is {mean_temp_ny:.2f}")
```

#### Which states and cities have a temperature above 15 degrees Celsius?

```python
# your code here
temp[temp['Temperature'] > 15][['State','City']]
```

#### Now, return only the cities that have a temperature above 15 degrees Celsius.

```python
# your code here
warm_cities = temp[temp['Temperature'] > 15]['City']
print("The cities that have a temperature above 15ºC are:")
for city in warm_cities:
    print("-", city)
```

#### Which cities have a temperature above 15 degrees Celcius and below 20 degrees Celsius?

**Hint**: First, write the condition. Then, select the rows.

```python
# your code here

warm_cities_1 = temp[(temp['Temperature'] > 15)&(temp['Temperature'] <20)]['City']

print("The cities that have temperatures above 15ºC, and below 20ºC are:")
for city in warm_cities_1:
    print("-", city)
```

#### Find the mean and standard deviation of the temperature of each state.

```python
# your code here

temp_stats_state = temp.groupby('State')['Temperature'].agg(['mean','std'])
temp_stats_state
```

#### What's the average salary in this company?

#### Find the employees whose title is associate and whose salary is above 55.

```python
# your code here
rich_associates = employees[(employees['Title']=='associate')&(employees['Salary']>55)]['Name']
print("Associates with salary above $55")
print("-------------------------------------------------")
for name in rich_associates:
    print("-",name)
```

#### Group the employees by number of years of employment. What are the average salaries in each group?

```python
# your code here
avg_salaries_per_year = employees.groupby('Years')['Salary'].mean()
print("Average Salaries by Years of Employment:")
print("----------------------------------------")
for year, salary in avg_salaries_per_year.items():
    print(f"{year} year(s): ${salary:,.2f}")
```

####  What is the average salary per title?

```python
# your code here
avg_salary_per_title = employees.groupby('Title')['Salary'].mean()
print("Average Salaries by Title:")
print("--------------------------")
for title, salary in avg_salary_per_title.items():
    print(f"{title}: ${salary:,.2f}")
```

####  Find the salary quartiles.


```python
# your code here
quartiles = employees['Salary'].quantile([0.25,0.5,0.75])
print("Salary Quartiles:")
print("-----------------")
for quantile, value in quartiles.items():
    print(f"{quantile * 100:.0f}th percentile: ${value:,.2f}")
```

#### Is the mean salary different per gender?

```python
# your code here
avg_sal_gender = employees.groupby('Gender')['Salary'].mean()
print("Average Salaries by gender:")
print("---------------------------")
for gender, salary in avg_sal_gender.items():
    print(f"{gender}: ${salary:,.2f}")
    
"""
Yes, it seems that the mean salary is slightly higer for men, compared to average salary for women.
"""
```

#### Find the minimum, mean and maximum of all numeric columns for each company department.



```python
# your code here
stats_per_dep = employees.groupby('Department')[['Years','Salary']].agg(['mean','max'])
stats_per_dep
```

#### Bonus Question:  for each department, compute the difference between the maximum and the minimum salary.
**Hint**: try using `agg` or `apply` combined with `lambda` functions.

```python
# your code here
sal_range_dep = employees.groupby('Department')['Salary'].agg(lambda x: x.max()-x.min())
print("Salary Range by Department:")
print("---------------------------")
for department, range_value in sal_range_dep.items():
    print(f"{department}: ${range_value}")
```

#### Select all the customers who have bought more than 50 items.
**Hint**: Remember that you are not asked to find orders with more than 50 items but customers who bought more than 50 items. A customer with two orders of 30 items each should appear in the selection.

```python
# your code here
customer_items = orders.groupby('CustomerID')['Quantity'].sum()
customers_over_50_items = customer_items[customer_items > 50]

print("Customers who have bought more than 50 items:")
print(customers_over_50_items.index.tolist())

customers_over_50_items.count()
```

#### Select orders from Spain that include more than 50 items.

```python
# your code here
# First, filter the DataFrame for orders from Spain
spanish_orders = orders[orders['Country']=='Spain']
spanish_orders

# Group by 'InvoiceNo' to aggregate the quantities per order
spanish_order_quantities = spanish_orders.groupby('InvoiceNo')['Quantity'].sum()

# Filter these grouped data to find orders where the total quantity is greater than 50
spanish_orders_over_50_items = spanish_order_quantities[spanish_order_quantities>50]

# Now, to get the full details of these orders, merge this filtered information back to the original dataframe
detailed_spanish_orders_over_50 = spanish_orders[spanish_orders['InvoiceNo'].isin(spanish_orders_over_50_items.index)]

display(detailed_spanish_orders_over_50)
```

#### Show the frequency distribution of the amount spent in Spain.

```python
# your code here
spanish_amount_spent_freq = orders[orders['Country']== 'Spain']['amount_spent'].value_counts().sort_index()
display(spanish_amount_spent_freq)

plt.figure(figsize=(12, 6))  # Set a larger figure size for better readability
spanish_amount_spent_freq.plot(kind='line')  # Change to line plot
plt.title('Frequency Distribution of Amount Spent for Orders from Spain')
plt.xlabel('Amount Spent')
plt.ylabel('Frequency')
plt.xticks(rotation=45)  # Rotate x-axis labels to avoid clutter
plt.grid(True)  #  add a grid for easier reading
plt.show()
```
